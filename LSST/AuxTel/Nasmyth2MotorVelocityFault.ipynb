```python
import matplotlib
%matplotlib widget
from matplotlib import pylab as plt
from astropy.time import Time, TimeDelta

from bokeh.plotting import figure, output_notebook, show
from bokeh.models import LinearAxis, Range1d
output_notebook()

from lsst_efd_client import EfdClient
```

```python
client = EfdClient("summit_efd")
```

```python
start = Time("2021-02-17T03:08:21", format="isot", scale="tai")
end = Time("2021-02-17T03:08:22", format="isot", scale="tai")
```

```python
measured_ns2_vel = await client.select_packed_time_series('lsst.sal.ATMCS.measuredMotorVelocity', ['nasmyth2MotorVelocity'], start, end)
trajectory_ns2_vel = await client.select_packed_time_series('lsst.sal.ATMCS.trajectory', ['nasmyth2RotatorAngleVelocity'], start, end)
```

```python
p = figure(x_axis_type='datetime', plot_width=800, plot_height=400)
p.yaxis.axis_label = "Nasmyth 2 Velocity (degrees/sec)"
p.xaxis.axis_label = "Time"
p.line(x=measured_ns2_vel.index.values, y=measured_ns2_vel['nasmyth2MotorVelocity'], color='red', alpha=0.5, legend_label='Measured Nasmyth Velocity Motor 2')
p.line(x=trajectory_ns2_vel.index.values, y=trajectory_ns2_vel['nasmyth2RotatorAngleVelocity'], color='blue', alpha=0.5, legend_label='Trajectory Nasmyth Velocity Motor 2')
p.legend.location = 'bottom_left'
p.legend.click_policy = 'hide'
show(p)
```
