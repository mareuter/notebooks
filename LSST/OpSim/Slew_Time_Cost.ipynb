```python
import matplotlib.pylab as plt
import numpy
%matplotlib inline
```

```python
def slew_cost(slew_time, ref_time, cost_ref, max_time):
    cost_dc = cost_ref * (max_time - ref_time) / (ref_time - cost_ref * max_time)
    cost_dt = -max_time * (cost_dc + 1)
    cost_k = cost_dc * cost_dt
    
    return ((cost_k / (slew_time + cost_dt)) - cost_dc - cost_ref) / (1.0 - cost_ref)
```

```python
slew_time = numpy.arange(0.0, 170.0, 0.1)
```

Plot variations of the slew cost curve keeping the maximum time and reference time the same as the current defaults.

```python
time_maximum = 150.0
time_reference = 5.0
cost_references = [0.05, 0.08, 0.1, 0.2, 0.3, 0.5]
for cost_reference in cost_references:
    cost_curve = slew_cost(slew_time, time_reference, cost_reference, time_maximum)
    plt.plot(slew_time, cost_curve, label="{}".format(cost_reference))
plt.xlabel("Slew Time (sec)")
plt.ylabel("Slew Cost")
plt.legend()
```
