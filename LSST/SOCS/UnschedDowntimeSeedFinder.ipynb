```python
from __future__ import division
from lsst.sims.ocs.downtime.unscheduled_downtime import UnscheduledDowntime
import numpy as np
```

```python
# Twenty year survey in days
survey_length = 7300
do_random = True
```

```python
#start_seed = 1618787900
#end_seed = 1618788000
start_seed = 1600000000
end_seed = 1630000000
for seed in xrange(start_seed, end_seed, 10):
    if seed % 100000 == 0:
        print("Trying Seed: {}".format(seed))
    ud = UnscheduledDowntime()
    ud.initialize(use_random_seed=do_random, random_seed=seed)
    num_downtimes = len(ud)
    if num_downtimes < 155 or num_downtimes > 160:
        continue
    down_days = ud.total_downtime
    if down_days < 320 or down_days > 325:
        continue
    percent_survey = down_days / survey_length * 100
    if percent_survey < 4.3 or percent_survey > 4.5:
        continue
    frac_days = down_days / num_downtimes
    if frac_days < 2.037 or frac_days > 2.038:
        continue
    total_c1 = 0
    total_c2 = 0
    for (n, duration, event_type) in ud.downtimes:
        if event_type.startswith("catastrophic"):
            if n < survey_length / 2 + 1:
                total_c1 += 1
            else:
                total_c2 += 1
    if total_c1 != 1 or total_c2 != 1:
        continue
    total_m1 = 0
    total_m2 = 0
    for (n, duration, event_type) in ud.downtimes:
        if event_type.startswith("major"):
            if n < survey_length / 2 + 1:
                total_m1 += 1
            else:
                total_m2 += 1
    if total_m1 + total_m2 < 10:
        continue
    if (total_m1 < 5 or total_m1 > 6) or (total_m2 < 5 or total_m2 > 6):
        continue   
    print("Seed: {}, Num Down: {}, Down Days: {}, Percent Survey: {:.1f}, Fraction Days: {:.3f}".format(seed, num_downtimes, down_days, percent_survey, frac_days))
    print("  Catastrophic Days: {}, {}. Major Days: {}, {}".format(total_c1, total_c2, total_m1, total_m2))
```
