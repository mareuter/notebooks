```python
from astropy.io import fits
import astropy.units as u
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
hdu = fits.open('StandardExposureWhiteLight1528409055.386417.fits')
```

```python
hdu[0].header
```

```python
data = hdu[0].data
```

```python
# Wavelength is in first block
wavelength = data[:, 0] * u.nm
# Intensity is in second block
intensity = data[:, 1]
```

```python
plt.figure()
plt.plot(wavelength, intensity)
plt.xlabel(r'$\lambda$ (nm)')
plt.ylabel(r'Intensity')
plt.show()
```

```python
cut_wave = np.logical_and(wavelength > 420 * u.nm, wavelength < 480 * u.nm)
```

```python
plt.figure()
plt.plot(wavelength[cut_wave], intensity[cut_wave])
plt.xlabel(r'$\lambda$ (nm)')
plt.ylabel(r'Intensity')
plt.show()
```
