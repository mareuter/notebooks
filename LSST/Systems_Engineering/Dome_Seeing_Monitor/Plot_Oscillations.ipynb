```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
change = 5
delta_time = 200
amplitude = 10
samples = 1000
point = 250
```

```python
x = np.arange(samples)
```

```python
y = point + (amplitude * np.sin(change * (x / delta_time)))
```

```python
plt.plot(x, y)
```
