```python
import matplotlib.pylab as plt
import numpy as np
import palpy
%matplotlib inline
```

```python
altitude = np.arange(5.0, 90., 0.1)
airmass1 = 1.0 / np.cos(np.radians(90.0 - altitude))
zenith_distance = 90.0 - altitude
airmass2 = palpy.airmasVector(np.radians(zenith_distance))
```

```python
plt.plot(altitude, airmass1, label="Analytic")
plt.plot(altitude, airmass2, label="PalPy")
plt.legend()
```

```python
delta_airmass = airmass2 - airmass1
```

```python
plt.plot(altitude, delta_airmass)
plt.ylim(-0.01, 0.001)
```
