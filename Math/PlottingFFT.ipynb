```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
Fs = 150.0
Ts = 1 / Fs
t = np.arange(0, 1, Ts)
```

```python
ff = 5
y = np.sin(2 * np.pi * ff * t)
```

```python
n = len(y)
k = np.arange(n)
T = n  / Fs
frq = k / T
#frq = frq[range(n // 2)]
```

```python
Y = np.fft.fft(y)
#Y = Y[range(n // 2)]
```

```python
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.plot(frq, abs(Y), '*r')
plt.xlabel('Freq (Hz)')
plt.ylabel('|Y(freq)')
plt.xlim(140,160)
```

```python
period = 1 / frq
```

```python
plt.plot(period, abs(Y), '*r')
plt.xlabel('Period (s)')
plt.ylabel('|Y(T)|')
```
